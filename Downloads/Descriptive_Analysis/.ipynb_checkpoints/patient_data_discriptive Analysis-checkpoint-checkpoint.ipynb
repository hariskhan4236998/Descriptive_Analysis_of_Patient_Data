```python
import pandas as pd 
```

```python
data = pd.read_csv("patient_diabetes_unclean.csv")
```

```python
data
```

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
print(data.describe())
```

Data Cleaning

```python
data.loc[:, 'Urea'] = data['Urea'].fillna(data['Urea'].mean())
```

```python
print(data['Gender'].value_counts())
print(data['CLASS'].value_counts())
```

```python
data = data.dropna()
```

```python
data.loc[:, 'Gender'] = data['Gender'].str.upper()
```

 Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Visualize Categorical Data

```python
sns.countplot(x='Gender', data=data)        #Bar plot for Gender distribution
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x='CLASS', data=data)        #Bar plot for CLASS distribution
plt.title('CLASS Distribution')
plt.show()
```

Visualize Numerical Data

```python
sns.boxplot(x='CLASS', y='AGE', data=data)       # Box plot for AGE grouped by CLASS
plt.title('Age Distribution by CLASS')
plt.show()
```

```python
data['BMI'].hist(bins=20)                  # Histogram for BMI
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

 Correlation Heatmap

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])
corr = numeric_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Advanced Insights

```python
sns.scatterplot(x='Urea', y='HbA1c', hue='CLASS', data=data)
plt.title('Relationship between Urea and HbA1c')
plt.show()
```

```python
sns.pairplot(data, hue='CLASS')
plt.show()
```

```python
data.to_csv('cleaned_data.csv', index=False)
```
